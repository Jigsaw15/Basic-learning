```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.columns
```

### Lineplot

```python
sns.lineplot(x = "sepal_length", y = "sepal_width", data = df) 
```

```python
sns.lineplot(x = "sepal_length", y = "sepal_width", data = df, hue = "species", style = "species", marker = 'o')
```

### Scatterplot

```python
sns.scatterplot(x = "sepal_length", y = "sepal_width", data = df, hue = "species")
```

### Bar plot

```python
sns.barplot(x = "sepal_length", y = "sepal_width", data = df)
```

```python
sns.barplot(x = "sepal_length", y = "sepal_width", data = df, orient = "h")
```

# Histogram

```python
sns.histplot(x = "sepal_length", data = df, kde = True)
```

```python
sns.displot(x = "sepal_length", data = df)
```

```python
sns.distplot(df["sepal_length"])
```

# Box Plot

```python
sns.boxplot(x = "sepal_length", data = df)
```

```python
sns.boxplot(x = "sepal_width", data = df)
```

# Join Plot

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = df)
```

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = df, hue = "species")
```

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = df, kind = "hex")
```

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = df, hue = "species")
```

# Reg Plot

```python
sns.regplot(x = "sepal_length", y = "sepal_width", data = df)
```

# Swarm Plot

```python
sns.swarmplot(x = "species", y = "sepal_length", data = df)
```

# pairplot

```python
sns.pairplot(data = df)
```

```python
sns.pairplot(x_vars=['sepal_length', 'sepal_width', 'petal_length'],y_vars="petal_width",data=df, )
```
