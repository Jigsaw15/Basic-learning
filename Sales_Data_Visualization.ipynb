### Exercise Question 3: Read motorcycle and classic car productline sales data and show
### it using a multiline plot and take first 100 rows from dataframe .
### Display the number of units sold per month for each product using multiline plots. (i.e., Separate
### Plotline for each product for each product).

```python
product = df.groupby("PRODUCTLINE")
motorcycle = product.get_group("Motorcycles")
classiccars = product.get_group("Classic Cars")
x = motorcycle["MONTH_ID"].head(100)
y = motorcycle["SALES"].head(100)
x1 = classiccars["MONTH_ID"].head(100)
y1 = classiccars["SALES"].head(100)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.plot(x, y, color = "b", marker = "o")
plt.plot(x1, y1,  color = "y", marker = "o")
```

##### Exercise Question 4: Read Motorcycles sales data of each month and show it using a scatter plot and take first 100 rows from data frame.

```python
product = df.groupby("PRODUCTLINE")
motorcycle = product.get_group("Motorcycles").head(100)
x = motorcycle["MONTH_ID"].head(100)
y = motorcycle["SALES"].head(100)
plt.scatter(x, y, label = "Sales data of each month")
plt.xlabel("No. of Months")
plt.ylabel("Motorcycles sold")
plt.grid(True, linewidth = 1, linestyle = "--")
plt.show()
```

#### Exercise Question 5: Read Classic Cars and motorcycle data product sales data and show it using the bar chart and take first 100 rows from data frame.

```python
product = df.groupby("PRODUCTLINE")
months = df["MONTH_ID"].head(100)
motorcycle = product.get_group("Motorcycles").head(100)
classiccar = product.get_group("Classic Cars").head(100)
x = motorcycle["SALES"]
x1 = classiccar["SALES"]
plt.bar([a-0.25 for a in months], x, width= 0.25, label = "Motorcycle data", align="edge")
plt.bar([a+0.25 for a in months], x1, width= -0.25, label = "CLassic cars data", align="edge")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.grid(True, linewidth = 1, linestyle = "--")
```

##### Exercise Question 8: Calculate total sale data for motor cycle and classic car product and show it using a Pie chart take first 100 rows from dataframe.

```python
product = df.head(100)
product = df.groupby("PRODUCTLINE")
bike = product.get_group("Motorcycles")
x1 = int(bike["SALES"].sum())
cars = product.get_group("Classic Cars")
x2 = int(cars["SALES"].sum())
plt.figure(figsize=(16,8))
plt.pie([x1,x2],labels=["Motorcycle","Classic Cars",],autopct="%1.1f%%")
plt.legend()
```

#### Exercise Question 9: Read motorcycle and classic car of all months and display it using the Subplot and take first 100 rows from data frame.

```python
product = df.groupby("PRODUCTLINE")
bike = product.get_group("Motorcycles")
cars = product.get_group("Classic Cars")
m1 = bike["MONTH_ID"].head(100)
m2 = cars["MONTH_ID"].head(100)
s1 = bike["SALES"].head(100)
s2 = cars["SALES"].head(100)
f, axarr = plt.subplots(2)
axarr[0].plot(m1, s1, label = "Bikes data", color="b", marker='o', linewidth=1)
axarr[0].set_title("Data of a Motorcycle")
axarr[1].plot(m2, s2, label = "Cars data", color="g", marker="o", linewidth=1)
axarr[1].set_title('Data of a car')
```

#### Exercise Question 10: Calculate total sale data for motor cycle ,classic car, truck and bus, vintage car product and show it using a Pie chart take first 1000 rows from data-frame.

```python
product = df.head(100)
product = df.groupby("PRODUCTLINE")
bike = product.get_group("Motorcycles")
x1 = int(bike["SALES"].sum())
cars = product.get_group("Classic Cars")
x2 = int(cars["SALES"].sum())
heavy = product.get_group("Trucks and Buses")
x3 = int(heavy["SALES"].sum())
vintage = product.get_group("Vintage Cars")
x4 = int(vintage["SALES"].sum())
plt.figure(figsize=(16,8))
plt.pie([x1,x2,x3,x4],labels=["Motorcycle","Classic","Trucks and Buses","Vintage Cars"],autopct="%1.1f%%")
plt.legend()
```
