```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../Downloads/loan_data.csv")
df
```

```python
df.isna().sum()
```

```python
df["Gender"] = df["Gender"].fillna("NA")
```

```python
df[["Married", "Dependents", "Self_Employed"]] = df[["Married", "Dependents", "Self_Employed"]].fillna("NA")
```

```python
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mode())
```

```python
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].mean())
```

```python
df["Credit_History"] = df["Credit_History"].fillna(0)
```
