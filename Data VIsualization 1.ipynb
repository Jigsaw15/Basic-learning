```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [4, 5 ,6 ,7, 1]
```

### Line PLot

```python
#marker = o, +, *, ^
#color = r, g, b k, c, m, y
plt.plot(x, y, marker = "o", color = "c", linestyle = ":", label = "my first line")
plt.xlabel("My x axis")
plt.ylabel("My y axis")
plt.xticks([1, 2, 3, 4, 5], ["a", "b", "c", "d", "e"])
plt.yticks([1, 2, 3, 4, 5, 6, 7], ["a", "b", "c", "d", "e", "g", "h"])
plt.title("My First lineplot")
plt.legend(loc = "best")
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,marker="o",color="r",linestyle="-." ,label="my first line")
plt.plot(y,x,marker="*",color="g",label="my second line")
plt.legend()
```

### Scatter Plot

```python
plt. scatter(x, y, s = 100, color = "r", label = "plt 1")
plt. scatter(y, x, s = 100, color = "y", label = "plt 1")
plt.legend()
```

### Bar plot

```python
plt.bar(x, y, color =["y", "g"])
plt.xlabel("my x axis")
```

### Histogram

```python
x = [1, 1, 2, 3,  4, 5]
```

```python
plt.hist(x)
plt.xlabel("Element")
plt.ylabel("Frequency")
plt.figure(figsize = (10, 10))
```

```python
plt.boxplot(x, y)
```

### Violin Plot

```python
plt.violinplot(x)
```

### Pie Chart

```python
plt.pie([20,80],labels=["a","b"],autopct="%1.3f%%", shadow = "True")
```
